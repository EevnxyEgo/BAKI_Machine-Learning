```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Load the training dataset
train_df = pd.read_csv("./train.csv")  
test_df = pd.read_csv("./test.csv") 
```

```python
X_train = train_df.drop("label", axis=1)  # features
y_train = train_df["label"]               # labels
```

```python
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
```

```python
# Standardize the feature data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
# Preprocessing Test Data
X_test = test_df.drop("label", axis=1)  # features
y_test = test_df["label"]               # labels
```

```python
# Encode labels in test data
y_test = label_encoder.transform(y_test)

# Standardize the test feature data
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd


# Mendapatkan nilai mean dan std
mean = scaler.mean_
std = np.sqrt(scaler.var_)

# Cetak atau simpan hasil
print("Mean:", mean)
print("Standard Deviation:", std)
```

```python
# Memformat nilai mean dan std dengan koma sebagai pembatas ribuan
mean_formatted = [f"{num:,.8f}" for num in mean]
std_formatted = [f"{num:,.8f}" for num in std]

# Menggabungkan elemen list menjadi string tanpa tanda petik
mean_str = ", ".join(mean_formatted)
std_str = ", ".join(std_formatted)

# Cetak hasil
print("Mean:", mean_str)
print("Standard Deviation:", std_str)
```

```python
# Define the model architecture
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')  # 3 classes (C, H, L)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_scaled, y_train, validation_split=0.2, epochs=10, batch_size=32)
```

```python
# Evaluate the model on test data
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
# Export the model in SavedModel format for TFLite conversion
model.export("plank_posture_model")

# Convert the model to TFLite format
converter = tf.lite.TFLiteConverter.from_saved_model("plank_posture_model")
tflite_model = converter.convert()

# Save the TFLite model
with open("plank_posture_model.tflite", "wb") as f:
    f.write(tflite_model)

print("Model converted to TFLite and saved as plank_posture_model.tflite")
```

```python
# Load TFLite model and allocate tensors
interpreter = tf.lite.Interpreter(model_path="plank_posture_model.tflite")
interpreter.allocate_tensors()

# Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()
```

```python
# Load input scaler
with open("./model/input_scaler.pkl", "rb") as f:
    input_scaler = pickle.load(f)

# Transform prediction into class
def get_class(prediction):
    return {
        0: "C",  # Correct
        1: "H",  # High Back
        2: "L",  # Low Back
    }.get(prediction)
```

```python
# Initialize MediaPipe pose
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
```

```python
# Determine important landmarks for plank
IMPORTANT_LMS = [
    "NOSE",
    "LEFT_SHOULDER",
    "RIGHT_SHOULDER",
    "LEFT_ELBOW",
    "RIGHT_ELBOW",
    "LEFT_WRIST",
    "RIGHT_WRIST",
    "LEFT_HIP",
    "RIGHT_HIP",
    "LEFT_KNEE",
    "RIGHT_KNEE",
    "LEFT_ANKLE",
    "RIGHT_ANKLE",
    "LEFT_HEEL",
    "RIGHT_HEEL",
    "LEFT_FOOT_INDEX",
    "RIGHT_FOOT_INDEX",
]

# Generate all columns of the data frame

HEADERS = ["label"] # Label column

for lm in IMPORTANT_LMS:
    HEADERS += [f"{lm.lower()}_x", f"{lm.lower()}_y", f"{lm.lower()}_z", f"{lm.lower()}_v"]
```

```python
VIDEO_TEST = "../../demo/plank2.mp4"
```

```python
# Capture video
cap = cv2.VideoCapture(VIDEO_TEST)
current_stage = ""
prediction_probability_threshold = 0.6

def draw_modern_status_box(image, current_stage, probability):
    overlay = image.copy()
    output = image.copy()
    cv2.rectangle(overlay, (0, 0), (300, 100), (30, 30, 30), -1)
    alpha = 0.7
    cv2.addWeighted(overlay, alpha, output, 1 - alpha, 0, output)
    cv2.rectangle(output, (0, 0), (300, 100), (10, 10, 10), thickness=2)
    cv2.putText(output, "STATUS", (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)
    cv2.putText(output, current_stage, (20, 70), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (100, 255, 100), 2, cv2.LINE_AA)
    cv2.putText(output, "PROB", (200, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)
    cv2.putText(output, str(round(probability, 2)), (200, 70), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (100, 100, 255), 2, cv2.LINE_AA)
    return output

def extract_important_keypoints(results):
    keypoints = []
    for lm in IMPORTANT_LMS:
        part = getattr(results.pose_landmarks.landmark, lm.lower())
        keypoints.extend([part.x, part.y, part.z, part.visibility])
    return keypoints

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, image = cap.read()
        if not ret:
            break
        image = cv2.resize(image, (640, 480))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = pose.process(image)
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if not results.pose_landmarks:
            print("No human found")
            continue

        # Draw landmarks with a modern style
        mp_drawing.draw_landmarks(
            image, 
            results.pose_landmarks, 
            mp_pose.POSE_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(100, 200, 255), thickness=2, circle_radius=2),  
            mp_drawing.DrawingSpec(color=(255, 150, 150), thickness=2, circle_radius=1)
        )

        try:
            # Extract keypoints from frame for the input
            row = extract_important_keypoints(results)
            X = np.array([row])
            X = input_scaler.transform(X)

            # Set the tensor to input the image to the model
            interpreter.set_tensor(input_details[0]['index'], X.astype(np.float32))
            interpreter.invoke()
            predictions = interpreter.get_tensor(output_details[0]['index'])[0]

            # Get predicted class and probability
            predicted_class = np.argmax(predictions)
            predicted_class_label = get_class(predicted_class)
            prediction_probability = predictions[predicted_class]

            # Evaluate model prediction
            if predicted_class_label == "C" and prediction_probability >= prediction_probability_threshold:
                current_stage = "Correct"
            elif predicted_class_label == "L" and prediction_probability >= prediction_probability_threshold: 
                current_stage = "Low back"
            elif predicted_class_label == "H" and prediction_probability >= prediction_probability_threshold: 
                current_stage = "High back"
            else:
                current_stage = "Unknown"

            # Apply modern status box design
            image = draw_modern_status_box(image, current_stage, prediction_probability)

        except Exception as e:
            print(f"Error: {e}")

        # Resize and display image
        resized_image = cv2.resize(image, (800, 600))
        cv2.imshow("CV2", resized_image)

        # Press Q to close cv2 window
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    for i in range(1, 5):
        cv2.waitKey(1)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import tensorflow as tf
import pickle
from sklearn.preprocessing import StandardScaler

# Load TFLite model and allocate tensors
interpreter = tf.lite.Interpreter(model_path="plank_posture_model.tflite")
interpreter.allocate_tensors()

# Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Load input scaler
with open("./model/input_scaler.pkl", "rb") as f:
    input_scaler = pickle.load(f)

# Function to transform prediction into class
def get_class(prediction):
    return {0: "C", 1: "H", 2: "L"}.get(prediction)

# MediaPipe initialization
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# Video capture
cap = cv2.VideoCapture("../../demo/plank2.mp4")
current_stage = ""
prediction_probability_threshold = 0.6

def draw_modern_status_box(image, current_stage, probability):
    overlay = image.copy()
    output = image.copy()
    cv2.rectangle(overlay, (0, 0), (300, 100), (30, 30, 30), -1)
    alpha = 0.7
    cv2.addWeighted(overlay, alpha, output, 1 - alpha, 0, output)
    cv2.rectangle(output, (0, 0), (300, 100), (10, 10, 10), thickness=2)
    cv2.putText(output, "STATUS", (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)
    cv2.putText(output, current_stage, (20, 70), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (100, 255, 100), 2, cv2.LINE_AA)
    cv2.putText(output, "PROB", (200, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)
    cv2.putText(output, str(round(probability, 2)), (200, 70), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (100, 100, 255), 2, cv2.LINE_AA)
    return output

def extract_important_keypoints(results):
    landmarks = results.pose_landmarks.landmark
    data = []
    for lm in IMPORTANT_LMS:
        keypoint = landmarks[mp_pose.PoseLandmark[lm].value]
        data.extend([keypoint.x, keypoint.y, keypoint.z, keypoint.visibility])
    return np.array(data).flatten().tolist()

# Process video frames
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, image = cap.read()
        if not ret:
            break
        image = cv2.resize(image, (640, 480))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = pose.process(image)
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if not results.pose_landmarks:
            print("No human found")
            continue

        # Draw landmarks
        mp_drawing.draw_landmarks(
            image, 
            results.pose_landmarks, 
            mp_pose.POSE_CONNECTIONS, 
            mp_drawing.DrawingSpec(color=(100, 200, 255), thickness=2, circle_radius=2),
            mp_drawing.DrawingSpec(color=(255, 150, 150), thickness=2, circle_radius=1)
        )

        try:
            # Extract keypoints and scale
            row = extract_important_keypoints(results)
            X = input_scaler.transform([row])

            # Set input tensor for model
            interpreter.set_tensor(input_details[0]['index'], X.astype(np.float32))
            interpreter.invoke()
            predictions = interpreter.get_tensor(output_details[0]['index'])[0]

            # Predicted class and probability
            predicted_class = np.argmax(predictions)
            predicted_class_label = get_class(predicted_class)
            prediction_probability = predictions[predicted_class]

            # Check class and probability threshold
            if predicted_class_label == "C" and prediction_probability >= prediction_probability_threshold:
                current_stage = "Correct"
            elif predicted_class_label == "L" and prediction_probability >= prediction_probability_threshold:
                current_stage = "Low back"
            elif predicted_class_label == "H" and prediction_probability >= prediction_probability_threshold:
                current_stage = "High back"
            else:
                current_stage = "Unknown"

            # Apply status box design
            image = draw_modern_status_box(image, current_stage, prediction_probability)

        except Exception as e:
            print(f"Error: {e}")

        # Display the image
        resized_image = cv2.resize(image, (800, 600))
        cv2.imshow("CV2", resized_image)

        # Press Q to exit
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    for i in range(1, 5):
        cv2.waitKey(1)
```
